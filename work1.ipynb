```python
import pandas as pd
import pymysql
```

```python
# 打开数据库连接
db = pymysql.connect(host="localhost",user="root",password="123456",database="work" )
```

```python
# 使用 cursor() 方法创建一个游标对象 cursor
cursor = db.cursor()
 
# 使用 execute() 方法执行 SQL，如果表存在则删除
cursor.execute("DROP TABLE IF EXISTS STUDENT")
 
# 使用预处理语句创建表
sql = """CREATE TABLE STUDENT (
         NAME  CHAR(20) NOT NULL,
         AGE INT,  
         SEX CHAR(1),
         STUDYNUMBER CHAR(12) )"""
 
cursor.execute(sql)
 
```

```python
# SQL 插入语句
sql = """INSERT INTO STUDENT(NAME,
        AGE, SEX, STUDYNUMBER)
         VALUES ('abc', 20, 'M', 201936661000)"""
try:
   # 执行sql语句
   cursor.execute(sql)
   # 提交到数据库执行
   db.commit()
except:
   # 如果发生错误则回滚
   db.rollback()
```
